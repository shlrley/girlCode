```python
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
cloth_df = pd.read_csv('../dataset/stfore_train.csv')
cloth_df.head()
```

```python
cust_df = pd.read_csv('../dataset/customer_data.csv')
cust_df.head()
```

```python
cloth_cust_df = pd.merge(cloth_df, cust_df, how='left', on=['external_code', 'retail'])
cloth_cust_df.head()
```

```python
cloth_cust_df = cloth_cust_df[['external_code', 'retail', 'category', 'color', 'fabric', 'customer']]
cloth_cust_df
```

```python
cloth_cust_df = cloth_cust_df.assign(product = cloth_cust_df.color.astype(str) + ' ' + \
    cloth_cust_df.fabric.astype(str) + ' ' + cloth_cust_df.category.astype(str))
cloth_cust_df.head()
```

```python
cloth_cust_df = cloth_cust_df.drop(columns=['category', 'color', 'fabric', 'external_code', 'retail'])
cloth_cust_df.head()
```

```python
cloth_cust_df = cloth_cust_df.dropna().sort_values('customer')
cloth_cust_df
```

```python
items = cloth_cust_df['product'].value_counts().sort_values(ascending=False)[:100].keys().tolist()
```

```python
# Tutorial: https://scrapfly.io/blog/web-scraping-with-selenium-and-python/
# configure webdriver
options = Options()
options.headless = True  # hide GUI
options.add_argument("--window-size=1920,1080")  # set window size to native GUI size
options.add_argument("start-maximized")  # ensure window is full-screen

chrome_options = webdriver.ChromeOptions()
chrome_options.add_experimental_option(
    # this will disable image loading
    "prefs", {"profile.managed_default_content_settings.images": 2}
)

driver = webdriver.Chrome(options=options, chrome_options=chrome_options)
driver.get('http://aritzia.com/en/clothing')

# Scroll down
driver.execute_script("window.scrollTo(0, 1080)") 

driver.find_element(By.XPATH, "//div[@class'='ar-viewmodes js-viewmodes df')]/button[contains(@class, 'ar-viewmodes__option js-viewmodes__option f0 bn black bg-transparent pa0 ml2-ns nowrap underline-hover ar-viewmodes__option--active')]/span[text()='4']").click()
print('Change to View 4')

# search = driver.find_element(By.CSS_SELECTOR, 'a.ws-normal ttc ttu-fr overflow-visible')
names = driver.find_elements(By.XPATH, "//div[contains(@class, 'f0 product-name ar-product-name js-product-plp-name pr4-ns')]")
# src = driver.find_elements(By.XPATH, "//div[@class='product-image js-product-plp-image tc  js-product-plp-image--trigger-qv]/a/img").__getattribute__('src')
src = driver.find_elements(By.XPATH, "//div//a//img[contains(@class, 'lazy w-auto w-100-l h-auto db') and starts-with(@src, 'https://aritzia.scene7.com/is/image/Aritzia/medium/')]")

links = []
for s in src:
    links.append(s.get_attribute("src"))
# driver.quit()
```

```python
names = [n.text for n in names]
names[:100]
```
